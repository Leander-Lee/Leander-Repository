```python
import requests
import time

results = []
for i in range(100):
    url = "https://api.wheretheiss.at/v1/satellites/25544"
    table = requests.get(url)
    if table.status_code == 200:
        data = table.json()
        timestamp = data["timestamp"]
        latitude = float(data["latitude"])
        longitude = float(data["longitude"])
        results.append((timestamp, latitude, longitude))
    else:
        Error = "Error"
        print(Error)
        break
    time.sleep(10)

print(results)
```

```python
#Could have already been done earlier in the code for better viewing but I preferred not to stop half way and just filter it

import matplotlib.pyplot as plt

time = [i[0] for i in results]
lat = [i[1] for i in results]
long = [i[2] for i in results]

#x-axis: longitude
#y-axis: latitude
plt.plot(long, lat, )

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Longitude vs Latitude")

plt.grid(True)

plt.show()
```
